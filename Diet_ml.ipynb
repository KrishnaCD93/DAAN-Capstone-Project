```python
def getTotals(df, group):
    dfTotal = df[[all([age, female, urban, edu]) 
    for age, female, urban, edu
    in zip(df.age==999, df.female==999, df.urban==999, df.edu==999)]]
    dfTotal = dfTotal[dfTotal['superregion2']=='HIC']
    dfTotal = dfTotal[dfTotal.columns.drop(['superregion2', 'age', 'female', 'urban', 'edu', 'median', 'upperci_95', 'lowerci_95', 's_lowerci_95', 's_upperci_95']+list(dfTotal.filter(regex='type')))]
    dfTotal['Group'] = group
    return dfTotal

FruitsTotals = getTotals(dfFruits, group = "Fruits Group")
NonStarchVegetablesTotals = getTotals(dfNonStarchVegetables, group = "Vegetables Group")
PotatoesTotals = getTotals(dfPotatoes, group = "Vegetables Group")
StarchyVegTotals = getTotals(dfStarchyVeg, group = "Vegetables Group")
BeanLegumeTotals = getTotals(dfBeanLegume, group = "Beans, Eggs, Meats, and Nuts Group")
NutSeedsTotals = getTotals(dfNutsSeeds, group = "Beans, Eggs, Meats, and Nuts Group")
RefinedGrainsTotals = getTotals(dfRefinedGrains, group = "Grains and Wheat Group")
WholeGrainsTotals = getTotals(dfWholeGrains, group = "Grains and Wheat Group")
ProcessedMeatTotals = getTotals(dfProcessedMeat, group = "Beans, Eggs, Meats, and Nuts Group")
SeafoodsTotals = getTotals(dfSeafoods, group = "Beans, Eggs, Meats, and Nuts Group")
EggsTotals = getTotals(dfEggs, group = "Beans, Eggs, Meats, and Nuts Group")
CheeseTotals = getTotals(dfCheese, group = "Cheese, Milk, and Yoghurt Group")
YoghurtTotals = getTotals(dfYoghurt, group = "Cheese, Milk, and Yoghurt Group")
MilkTotals = getTotals(dfMilk, group = "Cheese, Milk, and Yoghurt Group")

Totals = [FruitsTotals, NonStarchVegetablesTotals, PotatoesTotals, StarchyVegTotals,
    BeanLegumeTotals, NutSeedsTotals, RefinedGrainsTotals, WholeGrainsTotals,
    ProcessedMeatTotals, SeafoodsTotals, EggsTotals, CheeseTotals,
    YoghurtTotals, MilkTotals]
dfTotals = pd.concat(Totals)
dfTotals.dropna(inplace=True)
dfTotals.reset_index(inplace=True)
dfTotals.drop('index', axis=1, inplace=True)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error

seed = 18
```

```python
catFeaturesHigh = ['iso3']
catFeaturesLow = ['Group']

X = dfTotals.drop(['serving'], axis=1).copy()
y = dfTotals['serving'].copy()
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
categoricalTransformerHighCard = Pipeline(steps=[
    ('ordinal', OrdinalEncoder())
])
categoricalTransformerLowCard = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
preprocessor = ColumnTransformer(transformers=[
    ('cat_low_card', categoricalTransformerLowCard, catFeaturesLow),
    ('cat_high_card', categoricalTransformerHighCard, catFeaturesHigh)
], remainder='passthrough')
```

```python
models = {
    'XGBoost 1': XGBRegressor(
            booster = 'gbtree',
            eval_metric = "rmse",
            n_estimators = 10000,
            learning_rate = 0.035,
            reg_lambda = 0.5,
            reg_alpha = 24.5,
            subsample = 0.82,
            colsample_bytree = 0.13,
            max_depth = 10,
            tree_method='gpu_hist',
            predictor='gpu_predictor',
            gpu_id=0  
        ),

    'CatBoost': CatBoostRegressor(
            loss_function = 'RMSE',
            eval_metric = 'RMSE',
            depth = 4,
            l2_leaf_reg = 4.2,
            od_type = 'IncToDec',
            bagging_temperature = 290,
            iterations = 10000,
            thread_count = 4
        ),
    'SVM': SVR(), 'Random Forest': RandomForestRegressor()}
```

```python
my_pipeline = {'pipe1': Pipeline(steps=[
    ('preprocessor', preprocessor),
    (list(models)[0], models['XGBoost 1'])
]),
'pipe2': Pipeline(steps=[
    ('preprocessor', preprocessor),
    (list(models)[1], models['CatBoost'])
]),
'pipe3': Pipeline(steps=[
    ('preprocessor', preprocessor),
    (list(models)[2], models['SVM'])
]),
'pipe4': Pipeline(steps=[
    ('preprocessor', preprocessor),
    (list(models)[3], models['Random Forest'])
])}
```

```python
from tqdm import tqdm
predictions = {}
def getPreds(pipe):
    pipe.fit(X_train, y_train)
    return pipe.predict(X_test)

for i in tqdm(range(len(my_pipeline.keys())), desc='Running Models'):
    predictions[i] = getPreds(my_pipeline[f'pipe{i+1}'])
```

```python
for i in range(len(predictions.keys())):
    print(mean_squared_error(y_test, predictions[i], squared=False))
```
